## 2 state CTMC model - M/M/1/1 queue

```python
# import libraries
import numpy as np
from jmarkov.ctmc import ctmc
```

### Build the model

```python
# define model parameters
lda = 1
mu = 2 
```

```python
# build generator matrix Q
Q = np.array([[-lda, lda], [mu, -mu]])

mc = ctmc(Q)
print(f'Q={mc.generator}')
```

```python
# build initial probability vector
alpha = np.array([1, 0])
print(f'alpha={alpha}')
```

### Evaluate transient distribution

```python
t= 0.01
pi_t=mc.transient_probabilities(t, alpha)
print(f'pi({t})={pi_t}')
```
