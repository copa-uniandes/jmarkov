## CTMC M/M/1/K model

```python
# import libraries
import numpy as np
from jmarkov.ctmc import ctmc
```

### Build the model

```python
# define problem parameters
lda = 1
mu = 2 
K = 3
```

```python
# build generator matrix Q
Q = np.zeros((K+1,K+1))
for i in range(K+1):
    # arrivals 
    if i < K:
        Q[i,i+1] = lda
        Q[i,i] -= lda
    # services
    if i > 0:
        Q[i,i-1] = mu
        Q[i,i] -= mu

mc = ctmc(Q)
print(f'Q={mc.generator}')
```

```python
# build initial condition probability vector
alpha = np.zeros(K+1)
alpha[0] = 1
print(f'alpha={alpha}')
```

### Evaluate transient distribution

```python
t= 0.01
pi_t=mc.transient_probabilities(t, alpha)
print(f'pi({t})={pi_t}')
```
