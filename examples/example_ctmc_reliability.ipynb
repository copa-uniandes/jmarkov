## CTMC reliability model
An airplane has 4 engines. Each engine's uptime follows an exponential distribution with rate $\lambda$. The airplane needs at least one engine working. 

Model the airplane uptime as a CTMC.

```python
# import libraries
import numpy as np
from jmarkov.ctmc import ctmc
```

### Build the model

```python
# define problem parameters
lda = 0.001
K = 4
```

```python
# build generator matrix Q
Q = np.zeros((K+1,K+1))
for i in range(K+1):
    # failures 
    if i > 0:
        Q[i,i-1] = i*lda
        Q[i,i] -= i*lda

mc = ctmc(Q)
print(f'Q={mc.generator}')
```

```python
# build initial condition probability vector
alpha = np.zeros(K+1)
alpha[K] = 1
print(f'alpha={alpha}')
```

### Evaluate transient distribution

```python
t= 1
pi_t=mc.transient_probabilities(t, alpha)
print(f'pi({t})={pi_t}')
```
